```python
import numpy as np

M = np.array([[0.1,0.4,0.7,1], [0.3,0.7,0.2,0], [0.6,0.9,0.8,0], [0.5,0.7,0.1,1]])

d = (M[:,-1:]) # dentro da matriz M, está pegando a última coluna que é o nosso resultado desejado

# Organizando Matriz dados
T = np.zeros((M.shape[0], M.shape[1])) # cria matriz T, com tamanho de M.shape[0] - referente ao numero de linhas e M.shape[1] - ref ao numero de colunas
T[:,1:] = M[:,:-1] # sobrepõe os dados da M na T, sem a última coluna da M, que é o resultado esperado (0 ou 1) e a partir da coluna 2 da T, pois a coluna 1 é o limiar (-1)
T[:,0:1] = np.array([(-1*np.ones(M.shape[0])).tolist()]).T # sobrepõe a coluna 1 da T por -1. cria um array de ONES com o n de linhas da M, multiplica por -1, transforma em lista, depois em np.array e transpõe com .T  

# Peso sinaptico aleatorio
w = np.zeros(M.shape[0])
w[0] = -1
w[1:4] = np.random.rand(3)
wi = w; # peso inicial

print(f'w inicial: ' + str(wi))

n = 0.5 # estipulando valor de neta taxa de aprendizado
epoca = 0 # iniciando o contador de épocas

def FuncDegrau(q):
    g = np.zeros(len(q))
    for k in range(len(q)):
        if q[k] >= 0:
            g[k] = 1
        else:
            g[k] = 0
    return g

# Inicio da fase de treinamento
erro = True
while erro == True:

    erro = False

    epoca = epoca+1

    for k in range(M.shape[0]):
        x = T[k,:]
        u = [w @ x] # converte para array para comparar os comprimentos

        y= FuncDegrau(u)

        if y != d[k]:
            w = w + (n*(d[k]-y)*x).T
            erro = True
    print('Época:' + str(epoca) + '-> W: ' + str(w))
    if epoca == 1000: # Limite das épocas
        break # Força a saída do laço

print('w final: ' + str(w))
print('Época para convergência: ' + str(epoca))

u = np.array([w @ T]) # u será o vetor --> *w - aleatórios* transpostos
# vezes a linha desse vtor 'x' (treinamento)

out = np.zeros(u.shape[1])
print('Saída pré-rede: ' + str(u))
for k in range(u.shape[1]):
    out[k] = FuncDegrau(u[:,k]) # Aplicação da função bipolar (-1 1)

print('Resposta Rede ************** Desejado')
print([str(out) + '*****' + str(d.tolist())])    
```
