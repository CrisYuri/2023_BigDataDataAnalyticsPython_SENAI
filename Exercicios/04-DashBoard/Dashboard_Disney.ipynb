```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

# Criao do grfico de barras
categorias = df['listed_in'].value_counts()


plt.figure(figsize=(14,80))
plt.barh(categorias.index, categorias.values)

# Dados no eixo x
plt.yticks(df['listed_in'])

# Nome do eixo Y
plt.ylabel('Categoria')

# Nome do eixo X
plt.xlabel('Contagem')

#Ttlo do grfco
plt.title('Categorias')

# Apresentao da figura
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

# Criao do grfico de barras
content_per_year_type = df.groupby(['release_year', 'type']).size().unstack().fillna(0)
content_per_year_type

plt.figure(figsize=(14,6))
plt.bar(content_per_year_type['Movie'].index, content_per_year_type['Movie'].values, label='Movie')
plt.bar(content_per_year_type['TV Show'].index, content_per_year_type['TV Show'].values, label='TV Show')

# Dados no eixo x
plt.xticks(df['release_year'])

# Nome do eixo Y
plt.ylabel('Nmero de lanamentos')

# Nome do eixo X
plt.xlabel('Ano')

#Ttlo do grfco
plt.title('Nmero de lanamentos Disney por ano')
plt.xticks(rotation=90)  # Rotate x-axis labels to vertical
plt.tight_layout()
# Apresentao da figura
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('/Users/crisyuriok/Library/CloudStorage/OneDrive-Personal/1-BigDataDataAnalyticsPyhton/Exercicios/04-DashBoard/disney_plus_titles.csv',
                sep = ',', 
                decimal=',',
                encoding='utf-8')

content_rating = df['rating'].value_counts()
labels = df['rating'].value_counts().keys()

fig1, ax1 = plt.subplots(figsize=(6,6))

c = ['#A08F9D', '#A37E9D', '#A63B95', '#BB0FA0', '#de3163', '#00ffef', '#dbe852', '#a3c4eb', '#ff8243']

ax1.pie(content_rating, labels=labels ,autopct='%1.1f%%', startangle=90, colors=c)

plt.title('Indicao etria')

# plt.legend(('TV-G: General Audience\n', 'TV-PG: Parental Guidance\n','G: General Audience\n','PG-13: Parents Strongly Cautioned\n','TV-Y7-FV: Fantasy Violence'))


plt.show()
```

```python
content_type_dist = df['type'].value_counts()

plt.figure(figsize=(4,4))
content_type_dist.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Tipos de mdia')
plt.show()
```

```python
x = df['release_year']
y = df['duration']
plt.figure(figsize=(14,30))

plt.scatter(x, y , marker='*')
plt.title('Grfico de Disperso entre durao e ano')
plt.show()
```
